```python
# agg method
# passing a dictionary

genre.agg(
    {
        "Runtime":"mean",
        "IMDB_Rating":"mean",
        "No_of_Votes":"sum",
        "Gross":"sum",
        "Metascore":"max"
    }
)
```

```python
movies.groupby("Genre").agg({
        "Runtime":"mean",
        "IMDB_Rating":"mean",
        "No_of_Votes":"sum",
        "Gross":"sum",
        "Metascore":"max"
    })
```

```python
movies["Runtime"].agg(["min", "max", "mean"])
```

```python
genre.agg(['min','max','sum','mean'])
```

```python
genre.agg({
    "Runtime":['min', 'max','mean'],
    "IMDB_Rating": ['min', 'max', 'mean'],
    "No_of_Votes":["sum", 'max'],
    "Gross":['min','max'],
    "Metascore":['min','max']
})
```

```python
# looping on groups

df = pd.DataFrame(columns=movies.columns)

for group,data in genre:
  df = df.append(data[data["IMDB_Rating"]== data["IMDB_Rating"].max()])

df
```

```python
# split(apply) and combine
# apply --> builtin function

genre.apply(max)
```

```python
# number of movies starting with A in each group

def movies_start_with_A(group):
  return group["Series_Title"].str.startswith("A").sum()
```

```python
genre.apply(movies_start_with_A)
```

```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group["genre_rank"] = group["IMDB_Rating"].rank(ascending = False)
  return group
```

```python
genre.apply(rank_movie)
```
